## Test Tracking Data Box Score

```python
import json
```

```python
data = {}
with open("Basketball/Player_Tracking/CLEBOS/2017051902_nba-bos_BOX.json", "r") as read_file:
    data = json.load(read_file)
```

```python
keys = list(data.keys())
```

```python
summarized_data = {k: None for k in keys}
summarized_data
```

```python
for key in keys:
    if key == 'game-state':
        summarized_data[key] = pd.DataFrame(data[key], index=[0])
    else:
        summarized_data[key] = pd.DataFrame(data[key], index=[i for i in range(len(data[key]))])
```
